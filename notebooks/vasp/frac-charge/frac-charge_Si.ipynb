# Perform fractional charge test with pynter

Fractional charge test for Silicon. We will perform 6 calculations: from nelect to nelect +1 with interval 0.2.

```python
from pynter.tools.materials_project import MPDatabase
from pynter.vasp.schemes import Schemes
from pynter.data.datasets import Dataset
import os
homedir = os.getenv('HOME')

ds_path = os.path.join(homedir,'pynter-tutorials-files','frac-charge-Si')

# get structure of FCC Si from Materials Project
structure = MPDatabase(mp_id='mp-149').get_structure()

# Adapt job settings for a smaller job
job_settings = {'nodes':1,'timelimit':'00:10:00','add_automation':None}

# create scheme
schemes = Schemes(path=ds_path,structure=structure,job_settings=job_settings,name='Si-test')
jobs = schemes.fractional_charge_linearity()

ds = Dataset(jobs)
```

```python
ds.jobs_table()
```

```python
ds.write_jobs_input()
```

```python
ds.sync_dataset_to_hpc()
```

```python
for job in ds.jobs:
    job.run_job(sync=False)
```

```python
ds.queue()
```

```python
ds.sync_jobs()
```

```python
ds.jobs_table()
```
