```python
from pynter.data.datasets import Dataset
```

```python
ds = Dataset.from_json('./data/dataset.json')
```

```python
from pynter.tools.utils import set_display_df
set_display_df()
ds.jobs_table()
```

```python
job_bulk = ds.select_jobs(name='NN_Pbcm_2x2x1')[0]
jobs = ds.select_jobs(group='vacancies-PBE-2x2x1-supercell',common_node='2-PBE-OPT')
ds.jobs_table(jobs)
```

```python
from pynter.defects.entries import DefectEntry
entries = []
for j in jobs:
    data = {'stress':job_bulk.stress}
    entry = DefectEntry.from_jobs(j,job_bulk,{},multiplicity=None,data=data,tol=0.05)
    entries.append(entry)
```

```python
from pynter.defects.analysis import DefectsAnalysis
da = DefectsAnalysis(entries,job_bulk.vbm,job_bulk.energy_gap)
da.get_dataframe()
```

```python
from pynter.phase_diagram.chempots import Reservoirs
res = Reservoirs.from_json('./data/reservoirs.json')
res.get_dataframe()
```

```python
da.plot(res['X']);
```
