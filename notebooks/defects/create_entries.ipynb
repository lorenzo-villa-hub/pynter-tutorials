Select job relative to the bulk structure and relaxed defect calculations

```python
from pynter.defects.corrections import get_kumagai_correction_from_jobs
from pynter.defects.entries import DefectEntry
from pynter.defects.structure import defect_finder
```

```python
job_bulk = ds.select_jobs(name='NN_Pbcm_2x2x1')[0]
job_bulk
```

```python
jobs = ds.select_jobs(common_node='2-PBE-OPT')
ds.jobs_table(jobs,display=['final_energy'])
```

# Check automatic defect determination
To automate the creation of `Defect` objects, the `defect_finder` function is used. This performs a check that the defects are found correctly. If only the relaxed structure is available, you might have to play with the `tol` parameter, which defines the tolerance for site comparison in the relative defect and bulk structures. If the unrelaxed structure is provided, this is much easier, as the sites do not differ from the bulk sites.

```python
structure_bulk = job_bulk.final_structure
for j in jobs:
    defect = defect_finder(j.initial_structure,structure_bulk,tol=5e-02)
    print(defect)
```

# Create `DefectEntry` objects

Define the dielectric tensor, which is needed for the Kumagai corrections. If Freysoldt corrections are used, only the dielectric constant is needed.

```python
import numpy as np
dielectric_tensor = np.array([[11.28,0,0],[0,10.54,0],[0,0,11.96]])
```

Get kumagai corrections using `VaspJob` objects are inputs, and create a `DefectEntry` object for each DFT calculation. This object can be also instanciated manually, or starting from pymatgen `Structure` objects (needed in case a computational method that does not have a `Job` object was used, currectly only VASP is available).
Additional data (in this example residual stresses) can be included through the data dictionary.

```python
entries = []
for j in jobs:
    if j.group != 'defect-dipoles':
        kumagai_dict, plt = get_kumagai_correction_from_jobs(j,job_bulk,dielectric_tensor,tol=5e-02,get_plot=True)
        corrections = {'kumagai':sum([v for v in kumagai_dict.values()])}
        print(j.name,corrections)
        plt.ylim(-0.3,2)
        plt.show()
    else:
        corrections = {}
    data ={'stress':j.stress}
    entry = DefectEntry.from_jobs(job_defect=j,job_bulk=job_bulk,corrections=corrections,data=data,tol=5e-02)
    entries.append(entry)
```

# Create `DefectsAnalysis` object

This object is used to collectively analyse multiple defect calculations. A list of `DefectEntry` objects is used as input.

```python
from pynter.defects.analysis import DefectsAnalysis
```

```python
da = DefectsAnalysis(entries=entries,vbm=job_bulk.vbm,band_gap=job_bulk.energy_gap)
```

Multiplicities need to be manually defined. They can be also set after the object is created.

```python
multiplicities = {'Vac_Na':32,'Vac_Nb':32,'Vac_O':96,'Sub_Ca_on_Na':32,'Sub_Sr_on_Na':32,'Sub_Sn_on_Nb':32,
                  'Sub_Ca_on_Na-Vac_Na':32*4,'Sub_Sr_on_Na-Vac_Na':32*4,'Sub_Sn_on_Nb-Vac_O':32*6,'Vac_Na-Vac_O':32*12}
for entry in da:
    entry.multiplicity = multiplicities[entry.name]
```

The object can be visualized as a `pandas.DataFrame` object. Use the display argument to add columns based on methods available in the `DefectEntry` class.

```python
da.get_dataframe(display=['corrections'])
```

# Export `json` object

The object can be exported as `json`. This helps the analysis without having to re-import the DFT data every time and facilitates the data storage in a small file containing only necessary information.

```python
da.to_json('./data/defects_analysis.json')
```
